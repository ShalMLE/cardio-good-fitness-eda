```python
#Import the Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Make plots prettier
sns.set(style="whitegrid")
```

```python
#Load the Data
df = pd.read_csv('CardioGoodFitness.csv')

#view top rows
df.head()
```

# Data Overview

```python
#Shape of Data
print('Rows, Columns:',df.shape)

#Datatypes and Missing Values
df.info()
```

# Descriptive Statistics

```python
df.describe().transpose()
```

Insights:
1. Most customers earn around 53,000.
2. Income is between 29k to 104K.
3. 50% of the customers earn between 47k to 50k.

```python
#Null value check
df.isnull().sum()
```

# Univariate EDA with Statistics and Visuals

```python
#Age Distribution
df['Age'].describe()
sns.histplot(df['Age'],kde=True)
plt.title('Age Distribution');
```

```python
#Income Distribution
df['Income'].describe()
sns.histplot(df['Income'],kde=True)
plt.title('Income Distribution');
plt.xticks(rotation=45);
```

```python
#Product Preferences
sns.countplot(x='Product',data=df)
plt.title('Product Purchase Preference');
```

# Bivariate EDA with Statistics

```python
#Age vs Product
sns.boxplot(x='Product', y='Age', data=df)
plt.title('Age vs Product Preference');
```

```python
#Income vs Product
sns.boxplot(x='Product',y='Income',data=df)
plt.title('Income vs Product Preference');
```

```python
#Usage vs Fitness
sns.boxplot(x='Fitness',y='Usage',data=df)
plt.title('Weekly Usage by Fitness Level');
```

# Correlation

```python
df.corr()
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Matrix');
```

### Business Insights:

- Treadmill buyers are younger (~2530) and have higher income (~60K+).
- CrossTrainer and Stepper buyers are slightly older with moderate income.
- People with higher fitness scores dont necessarily use equipment more.
